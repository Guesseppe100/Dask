```python
# SEGMENTO 1 — Setup local notebook + Docker Dask workers

from pathlib import Path
import dask.dataframe as dd
from dask.distributed import Client

USE_SAMPLE_DATA = True

# Local paths: only for checking files from Windows
BASE_DIR_LOCAL = Path("..").resolve()

if USE_SAMPLE_DATA:
    DATA_DIR_LOCAL = BASE_DIR_LOCAL / "sample_data_demo"
    DATA_GLOB_DOCKER = "/home/jovyan/sample_data_demo/*.csv"
else:
    DATA_DIR_LOCAL = BASE_DIR_LOCAL / "data" / "secop_chunks"
    DATA_GLOB_DOCKER = "/home/jovyan/data/secop_chunks/*.csv"

OUTPUT_DIR_LOCAL = BASE_DIR_LOCAL / "output"
OUTPUT_DIR_LOCAL.mkdir(exist_ok=True)

print("LOCAL DATA_DIR:", DATA_DIR_LOCAL)
print("LOCAL OUTPUT_DIR:", OUTPUT_DIR_LOCAL)
print("CSV files local:", len(list(DATA_DIR_LOCAL.glob("*.csv"))))
print("DOCKER DATA GLOB:", DATA_GLOB_DOCKER)

client = Client("tcp://localhost:8786")
print("Workers:", client.scheduler_info()["workers"].keys())

ddf = dd.read_csv(
    DATA_GLOB_DOCKER,
    dtype=str,
    assume_missing=True,
    blocksize="64MB"
)

print("Rows:", ddf.shape[0].compute())
print("Partitions:", ddf.npartitions)
print("Columns:", list(ddf.columns))
```
